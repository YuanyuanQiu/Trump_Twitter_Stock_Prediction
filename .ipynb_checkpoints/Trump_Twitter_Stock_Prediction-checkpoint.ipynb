doc2vec Model

```python
# input tweets -- df.loc[:,['date','text']]
# output doc2vec model
def build_doc2vec_model(data, vec_size, alpha, max_epochs):
    tagged_data = [TaggedDocument(words=[w for w in word_tokenize(_d.lower()) if w not in stops],
                                  tags=[str(i)]) for i,_d in enumerate(data)]

    '''Note: dm defines the training algorithm. dm=1: distributed memory’ (PV-DM); dm =0:‘distributed bag of words’ (PV-DBOW).
    Distributed Memory model preserves the word order in a document
    whereas Distributed Bag of words just uses the bag of words approach, which doesn’t preserve any word order.'''
    model = Doc2Vec(vector_size=vec_size,
                    alpha=alpha, 
                    min_alpha=0.00025,
                    min_count=1,
                    dm =1)

    model.build_vocab(tagged_data)

    for epoch in range(max_epochs):
        print('iteration {0}'.format(epoch))
        model.train(tagged_data,
                    total_examples=model.corpus_count,
                    epochs=model.epochs)
        # decrease the learning rate
        model.alpha -= 0.0002
        # fix the learning rate, no decay
        model.min_alpha = model.alpha

#     model.save("sample.model")
#     print("Model Saved")
    return model
```

## Train/Test Split

```python
# input df -- df.loc[:,['date','text','Daily Return']]
# output dic -- {date: {'price': price, 'tweets': tweets, 'scores': scores}}
def get_tweet_and_stock_data(df):
    date_to_tweet = {}
    date_to_price = {}
    date_to_score = {}
    
    for index, row in df.iterrows():
        date = index
        
        if date not in date_to_tweet:
            date_to_tweet[date] = []
        date_to_tweet[date].append(row['text'])
        
        if date not in date_to_score:
            date_to_score[date] = []
        date_to_score[date].append(row['compound'])
        
        date_to_price[date] = row['Daily Return']
        
    composite_dict = {}
    lag = 0

    for date, tweets in date_to_tweet.items():
        date_adjusted = date + timedelta(days=lag)
        if date_adjusted not in date_to_price:
            continue
        composite_dict[date_adjusted] = {'price': date_to_price[date_adjusted], 'tweets': tweets, 'scores': np.mean(date_to_score[date])}
        
    return composite_dict
```

```python
composite_dict = get_tweet_and_stock_data(df)
```

```python
composite_dict
```

```python
# input composite_dic, d2vmodel
# output X, y, date_list
def get_xy_data(composite_dict, d2vmodel):
    X = []
    Y = []
    #date_to_embeddings = {}
    date_list = []
    for date, v in composite_dict.items():
        price = v['price']
        tweets = v['tweets']
        embedding_tweets = []
        
        for tweet in tweets:
            d2vmodel.random.seed(0)
            word_tokens = word_tokenize(tweet.lower())
            filtered_text = [w for w in word_tokens if w not in stops]
            embedding_tweets.append(d2vmodel.infer_vector(filtered_text))
            
        embedding_average = np.mean(embedding_tweets, axis=0)
        #date_to_embeddings[date] = embedding_average
        
        X.append(embedding_average)
        Y.append(price)
        date_list.append(date)

    return np.array(X), np.array(Y), date_list
```

```python
# Train
df_train = df.loc[train_start_date:train_end_date, :]
composite_dict_train = get_tweet_and_stock_data(df_train)
X_train, y_train, _ = get_xy_data(composite_dict_train, d2vmodel)
scaler = preprocessing.StandardScaler().fit(X_train)
X_train = scaler.transform(X_train)

# Test
df_test = df.loc[test_start_date:test_end_date, :]
composite_dict_test = get_tweet_and_stock_data(df_test)
X_test, y_test, date_list_test = get_xy_data(composite_dict_test, d2vmodel)
X_test = scaler.transform(X_test)
```

# Model

```python
def nn_model():
    model = keras.Sequential([
        keras.layers.Dense(128, activation='relu'),
        keras.layers.Dropout(0.5),
        keras.layers.Dense(64, activation='relu'),
        keras.layers.Dropout(0.5),
        keras.layers.Dense(1, activation=None)
    ])
    return model

def train_model(X_train, y_train, model, num_epochs, X_validation, y_validation):
    model.compile(optimizer='adam', loss='mean_squared_error', metrics=['mean_squared_error'])
    early_stop = EarlyStopping(monitor='val_loss', mode='min', verbose=0, patience=3)
    histories = model.fit(X_train, y_train, epochs=num_epochs, validation_data=(X_validation, y_validation),
                          batch_size=20, callbacks=[early_stop])
    return model, histories

def test_model(X_test, y_test, model):
    test_loss, test_acc = model.evaluate(X_test, y_test)
    print("Test loss:", test_acc)
    return test_acc

def plot_history(histories, key='loss'):
    plt.figure(figsize=(16,10))
    for name, history in histories:
        val = plt.plot(history.epoch, history.history['val_' + key],
                       '--', label=name.title()+' Val')
        plt.plot(history.epoch, history.history[key], color=val[0].get_color(),
                 label=name.title()+' Train')
    plt.xlabel('Epochs')
    plt.ylabel(key.replace('_',' ').title())
    plt.legend()
    plt.xlim([0, max(history.epoch)])
```

```python
date_to_embeddings_test = {}
for i in range(len(X_test)):
    date_to_embeddings_test[date_list_test[i]] = X_test[i]

num_epochs = 100
model = nn_model()
model, model_history = train_model(X_train, y_train, model, num_epochs, X_test, y_test)
model_acc = test_model(X_test, y_test, model)
```

```python
class MyStrategy(strategy.BacktestingStrategy):
    def __init__(self, feed, instrument, model_predictor, date_to_portfolio, start_date, date_to_embeddings_test):
        super(MyStrategy, self).__init__(feed, 10000)
        self.__position = None
        self.__instrument = instrument
        self.predictor = model_predictor
        # We'll use adjusted close values instead of regular close values.
        # self.setUseAdjustedValues(True)
        #self.__sma = ma.SMA(feed[instrument].getPriceDataSeries(), smaPeriod)
        self.results = date_to_portfolio
        self.start = start_date
        self.date_to_embeddings_test = date_to_embeddings_test

    def onEnterOk(self, position):
        execInfo = position.getEntryOrder().getExecutionInfo()
        self.info("BUY at $%.2f" % (execInfo.getPrice()))

    def onEnterCanceled(self, position):
        self.__position = None

    def onExitOk(self, position):
        execInfo = position.getExitOrder().getExecutionInfo()
        self.info("SELL at $%.2f" % (execInfo.getPrice()))
        self.__position = None

    def onExitCanceled(self, position):
        # If the exit was canceled, re-submit it.
        self.__position.exitMarket()

    def onBars(self, bars):
        # Wait for enough bars to be available to calculate a SMA.
        #if self.__sma[-1] is None:
            #return

        bar = bars[self.__instrument]
        if bar.getDateTime() < self.start:
            return
        if bar.getDateTime() not in self.date_to_embeddings_test:
            return 

        self.results[bar.getDateTime()] = self.getBroker().getEquity()
        percent_change = (self.predictor.predict(np.array([self.date_to_embeddings_test[bar.getDateTime()]])))[0][0]
        

        # If a position was not opened, check if we should enter a long position.
        if self.__position is None:
            #if bar.getPrice() > self.__sma[-1]:
            if percent_change > 0:
                # Enter a buy market order for 10 shares. The order is good till canceled.
                self.__position = self.enterLong(self.__instrument, 10, True)
        # Check if we have to exit the position.
        #elif bar.getPrice() < self.__sma[-1] and not self.__position.exitActive():
        elif percent_change < 0 and not self.__position.exitActive():
            self.__position.exitMarket()
```

```python
def stock_data_download(tickers, start, end):
    for ticker in tickers:
        stock = yf.download(ticker, start=start, end=end, interval = "1d", auto_adjust=False)
        stock_adj = yf.download(ticker, start=train_start_date, end=test_end_date, interval = "1d", auto_adjust=True)
        stock['Adj Close'] = stock_adj['Close']
        stock.to_csv('stock_data/' + ticker + ".csv")
```
